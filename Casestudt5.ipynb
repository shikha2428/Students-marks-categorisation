```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel("Marks_secC.xlsx", header=1)  # Skip first row (it had junk titles)

# Clean and rename columns (based on screenshot)
df.columns = ['Roll No', 'Name', 'Email', 'Course', 'Date', 'Status', 'Marks', 'Submitted']

# Calculate percentage
df['Percentage'] = (df['Marks'] / 15) * 100

# Categorize the students
above_75 = df[df['Percentage'] > 75]
between_60_75 = df[(df['Percentage'] >= 60) & (df['Percentage'] <= 75)]
below_60 = df[df['Percentage'] < 60]

# Print each category
print("Students with > 75%:")
print(above_75[['Roll No', 'Name', 'Percentage']], "\n")

print("Students with 60%-75%:")
print(between_60_75[['Roll No', 'Name', 'Percentage']], "\n")

print("Students with < 60%:")
print(below_60[['Roll No', 'Name', 'Percentage']], "\n")

# Plot Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Percentage'], bins=range(0, 101, 5), edgecolor='black', color='blue')
plt.title('Histogram Plot')
plt.xlabel('Percentage')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()

# Plot Scatter
plt.figure(figsize=(10, 5))
plt.scatter(df['Roll No'], df['Percentage'], color='blue')
plt.title('Scatter Plot')
plt.xlabel('Students')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
